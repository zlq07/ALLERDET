```python
import pandas as pd
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
```

```python
def txt_a_fasta(archivo_txt, archivo_fasta):
    with open(archivo_txt, 'r') as in_file, open(archivo_fasta, 'w') as out_file:
        lineas = in_file.readlines()
        
        # Eliminar espacios en blanco y saltos de línea
        lineas = [linea.strip() for linea in lineas if linea.strip()]

        # Iterar sobre líneas, asumiendo que cada encabezado está seguido por su secuencia
        for i in range(0, len(lineas), 2):
            header = lineas[i]
            secuencia = lineas[i + 1] if i + 1 < len(lineas) else ''  # Verificar que hay una secuencia
            
            # Cambiar encabezados según las reglas
            if header.startswith(">Negative_"):
                header = header.replace("Negative_", "non-allergen_")
            elif header.startswith(">Positive_"):
                header = header.replace("Positive_", "allergen_")
            
            # Escribir en formato FASTA
            out_file.write(f"{header}\n")  # Escribir el encabezado
            out_file.write(f"{secuencia}\n")  # Escribir la secuencia

# Especifica los nombres de los archivos
archivo_txt = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/SEP-AlgPro/merging_AlgPred2_NetAllergen.txt'    # Tu archivo de texto
archivo_fasta = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/SEPAlgPro.fasta' # Archivo de salida FASTA

txt_a_fasta(archivo_txt, archivo_fasta)
```

```python
from Bio import SeqIO

# Lista de archivos a combinar
archivo1 = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/PreAlgPro/train.fasta'
archivo2 = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/PreAlgPro/test.fasta'
archivo3 = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/SEPAlgPro.fasta'

archivos = [archivo1, archivo2, archivo3]
# Archivo de salida
salida = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/alldata.fasta'

with open(salida, 'w') as out_file:
    for archivo in archivos:
        for record in SeqIO.parse(archivo, "fasta"):
            SeqIO.write(record, out_file, "fasta")
```

```python
def eliminar_repetidos(archivo_fasta, archivo_salida):
    secuencias_vistas = set()  # Conjunto para almacenar secuencias únicas
    with open(archivo_fasta, 'r') as in_file, open(archivo_salida, 'w') as out_file:
        for record in SeqIO.parse(in_file, "fasta"):
            # Usar la secuencia como clave para evitar duplicados
            if str(record.seq) not in secuencias_vistas:
                secuencias_vistas.add(str(record.seq))  # Añadir secuencia al conjunto
                SeqIO.write(record, out_file, "fasta")  # Escribir registro único

# Especifica los nombres de los archivos
archivo_fasta = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/alldata.fasta'    # Tu archivo FASTA original
archivo_salida = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/alldata_u.fasta' # Archivo de salida sin duplicados

eliminar_repetidos(archivo_fasta, archivo_salida)
```

```python
def separar_fasta(archivo_fasta, archivo_allergen, archivo_non_allergen):
    # Contadores para las secuencias
    contador_allergen = 0
    contador_non_allergen = 0

    with open(archivo_fasta, 'r') as in_file, \
         open(archivo_allergen, 'w') as out_file_allergen, \
         open(archivo_non_allergen, 'w') as out_file_non_allergen:

        for record in SeqIO.parse(in_file, "fasta"):
            # Verificar el encabezado para clasificar
            if "non-allergen_" in record.id:
                SeqIO.write(record, out_file_non_allergen, "fasta")
                contador_allergen += 1
            elif "allergen_" in record.id:
                SeqIO.write(record, out_file_allergen, "fasta")
                contador_non_allergen += 1

    # Mostrar resultados
    print(f"Número de secuencias de alérgenos: {contador_allergen}")
    print(f"Número de secuencias de no alérgenos: {contador_non_allergen}")

# Especifica los nombres de los archivos
archivo_fasta = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/alldata_u.fasta'  # Tu archivo FASTA original
archivo_allergen = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/all_allergen.fasta'  # Archivo de salida para alérgenos
archivo_non_allergen = '/Users/z07/Desktop/TFG/ALLERDET/new_app/data/all_non_allergen.fasta'  # Archivo de salida para no alérgenos

separar_fasta(archivo_fasta, archivo_allergen, archivo_non_allergen)
```
